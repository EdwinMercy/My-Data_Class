```python
category_var=['surgery', 'age','temp_of_extremities', 'peripheral_pulse','mucous_membrane', 'capillary_refill_time', 
'pain','peristalsis','abdominal_distention', 'nasogastric_tube', 'nasogastric_reflux',
'rectal_exam_feces', 'abdomo_appearance','abdomen','surgical_lesion', 'cp_data']
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
for Car in category_var:
    df[Car]=enc.fit_transform(df[Car])
```

```python
df.transpose()
```

```python
X = df.drop('outcome', axis =1)
y = df['outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn = KNeighborsClassifier()
Knn.fit(X_train,y_train)
```

```python
Knn.score(X_test,y_test)
```

```python
pred =Knn.predict(X_test)
```

```python
pred
```

```python
score_r =[]
for i in range (1,100):
    Knn_r= KNeighborsClassifier()
    Knn_r.fit(X_train,y_train)
    score_r.append(Knn_r.score(X_test,y_test))
```

```python
plt.plot(range(1,100),score_r)
plt.scatter(range(1,100),score_r, color ='r')
plt.xlabel('Prediction Score')
plt.ylabel('Number of Neighbors')
plt.show()
```
