```python
import numpy as np
import pandas as pd
```

```python
rate = pd.read_csv('update_exchange_rate.csv')
```

```python
rate.head()
```

```python
rate.isnull().sum() #to check for missing values
```

```python
rate.rename({'Unnamed: 0':'m'}, axis =1, inplace = True)
```

```python
rate.head()
```

```python
rate.drop('m', axis= 1, inplace = True)
```

```python
rate.head()
```

```python
rate.info()
```

```python
rate['Rate Date'] = pd.to_datetime(rate['Rate Date'])
```

```python
rate.info()
```

```python
rate['Year'] = pd.DatetimeIndex(rate['Rate Date']).year
rate['Month'] = pd.DatetimeIndex(rate['Rate Date']).month
rate['Day'] = pd.DatetimeIndex(rate['Rate Date']).day
rate['Weekdays'] = pd.DatetimeIndex(rate['Rate Date']).dayofweek
```

```python
rate.tail()
```

```python
rate[(rate['Rate Date'] <= pd.Timestamp(2021,3,30)) &(rate['Rate Date'] >=pd.Timestamp(2021,2,1))]
```

```python
rate.to_excel('updated_exchange_rate.xlsx', sheet_name = 'exchange') #converting to excel single sheet
```

```python
with pd.ExcelWriter('Exchange_Rate102.xlsx', engine = 'xlsxwriter') as writer:  #converting to excel multiple sheet
        rate[rate['Currency']=='USD'].to_excel(writer, sheet_name = 'USD' ,index= False ,engine ='xlsxwriter')
        rate[rate['Currency']=='EURO'].to_excel(writer, sheet_name = 'EURO' ,index= False ,engine ='xlsxwriter')
        rate[rate['Currency']=='YEN'].to_excel(writer, sheet_name = 'YEN' ,index= False ,engine ='xlsxwriter')
        rate[rate['Currency']=='NAIRA'].to_excel(writer, sheet_name = 'NAIRA' ,index= False ,engine ='xlsxwriter')
```
