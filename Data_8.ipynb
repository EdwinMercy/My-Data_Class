```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
```

```python
tip
```

```python
sns.distplot(tip['total_bill'],kde = False)
```

```python
sns.jointplot(x ='total_bill',y ='tip',data = tip, kind ='reg')
```

```python
sns.regplot(x= tip['total_bill'],y=tip['tip'])# checks what influences the dependent varible
```

```python
sns.regplot(x= tip['total_bill'],y=tip['size'])
```

```python
sns.regplot(x= tip['size'],y=tip['tip'])
```

```python
sns.pairplot(tip,) #hue adds colors to the data provided
```

```python
#categorical data
```

```python
plt.figure(figsize =(12,6))
sns.barplot(x ='day', y = 'total_bill',data = tip, palette ='tab10',estimator =np.max)
plt.show()
```

```python
sns.countplot(x ='smoker',data = tip, hue='sex')
plt.show()
```

```python
sns.violinplot(x='time',y='tip',data =tip,hue = 'sex',dodge = True)
sns.swarmplot(x='time',y='tip',data =tip,hue = 'sex',dodge = True)
```

```python
sns.swarmplot(x='time',y='tip',data =tip,hue = 'sex',dodge = True)
```
